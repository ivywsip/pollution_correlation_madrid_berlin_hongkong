```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
from pandas.io.json import json_normalize
```

```python
url="https://public.opendatasoft.com/api/records/1.0/search/?dataset=statistisches-bundesamt-12411-0010&rows=1000&facet=lander&refine.lander=Berlin"
response=requests.get(url)
response
```

```python
r=response.json()
```

```python
serie=pd.DataFrame(r["records"])["fields"]
serie

ber_pop = json_normalize(serie)
ber_pop["date"]=pd.to_datetime(ber_pop["date"])
```

```python
ber_pop.sort_values(by="date",inplace=True)
```

```python
ber_pop.reset_index(drop=True,inplace=True)
```

```python
ber_pop
```

```python
# Formatting date and population (in million)
ber_pop['date'] = ber_pop['date'].dt.year

ber_pop
```

```python
# Adding missing date for 2018 

value_2018 = {'date':2018,
              'lander':'Berlin',
              'population':3748148}

ber_pop = ber_pop.append(value_2018,ignore_index=True,sort=True)

value_2019 = {'date':2019,
              'lander':'Berlin',
              'population':3754418}

ber_pop = ber_pop.append(value_2019,ignore_index=True,sort=True)
```

```python
# format Berlin to berlin 
ber_pop['lander'] = ber_pop['lander'].str.lower()


# rename column 
ber_pop = ber_pop.rename(columns={'lander':'city','date':'year'})
```

```python
ber_pop = ber_pop[['year','population','city']]
ber_pop
```

```python
ber_pop.to_pickle('./pickles/berlin_population.pkl')
```
