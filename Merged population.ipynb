# Combining population of three cities

```python
mad_pop = pd.read_pickle('./pickles/madrid_population.pkl')

mad_pop
```

```python
hk_pop = pd.read_pickle('././pickles/hk_population.pkl')
hk_pop
```

```python
ber_pop = pd.read_pickle('./pickles/berlin_population.pkl')
ber_pop
```

```python
all_pop = pd.concat([mad_pop,ber_pop,hk_pop]).reset_index(drop=True)
all_pop['year']=pd.to_datetime(all_pop['year'],format='%Y').dt.year
all_pop['population']=all_pop['population'].astype('int')
all_pop.groupby(['year','city']).sum()
```

```python
all_pop.to_pickle('./pickles/merged_population.pkl')
```
