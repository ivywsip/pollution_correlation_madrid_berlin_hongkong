```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://datosmacro.expansion.com/demografia/poblacion/espana-comunidades-autonomas/madrid"
```

```python
response=requests.get(url)
soup=BeautifulSoup(response.content,"html.parser")
soup
```

```python
table=soup.find_all("div",class_="table-responsive")
table[0]
```

```python
rows=table[0].find_all("tr")
rows_split=[row.text.strip().split("\n") for row in rows]
rows_split
```

```python
month=[]
date=[]
density=[]
men=[]
women=[]
population=[]

for r in rows_split[1:]:
    if r[0].startswith("Diciembre"):
        if len(r[0])>43:
            month.append(r[0].split(" ")[0])
            date.append(r[0].split(" ")[1][:4])
            density.append(r[0].split(" ")[1][4:7])
            men.append(r[0].split(" ")[1][7:16])
            women.append(r[0].split(" ")[1][16:25])
            population.append(r[0].split(" ")[1][25:])
        else:
            month.append(r[0].split(" ")[0])
            date.append(r[0].split(" ")[1][:4])
            density.append(0)
            men.append(r[0].split(" ")[1][4:13])
            women.append(r[0].split(" ")[1][13:22])
            population.append(r[0].split(" ")[1][22:])
    else:
        if len(r[0])>38:
            month.append(r[0].split(" ")[0])
            date.append(r[0].split(" ")[1][:4])
            density.append(r[0].split(" ")[1][4:7])
            men.append(r[0].split(" ")[1][7:16])
            women.append(r[0].split(" ")[1][16:25])
            population.append(r[0].split(" ")[1][25:])
        else:
            month.append(r[0].split(" ")[0])
            date.append(r[0].split(" ")[1][:4])
            density.append(0)
            men.append(r[0].split(" ")[1][4:13])
            women.append(r[0].split(" ")[1][13:22])
            population.append(r[0].split(" ")[1][22:])
        
     
```

```python
data={"month":month,
      "date":date,
      "density":density,
      "men":men,
      "women":women,
      "population":population}

mad_pop=pd.DataFrame(data=data)
mad_pop
```

```python
# Modify columns 
mad_pop = mad_pop[mad_pop['month'] !='Junio']
mad_pop = mad_pop.drop(['month','density','men','women'],axis=1).assign(city='madrid').reset_index(drop=True)
```

```python
# We need only data from 2008-2018
mad_pop = mad_pop.iloc[0:11]
```

```python
mad_pop['population'] = mad_pop['population'].str.replace('.','')
mad_pop = mad_pop.rename(columns={'date':'year'})
mad_pop
```

```python
mad_pop.to_pickle('./pickles/madrid_population.pkl')
```
